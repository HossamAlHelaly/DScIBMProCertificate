## Chapter 5. Data Visualization with NumPy and Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(5)
x
```

```python
y = 2 * x
y
```

```python
plt.plot(x, y, '--o')
```

### Single Line Plot:

```python
x = np.random.randint(1, 6, 6)
x
```

```python
plt.plot(x)
plt.xlabel('index')
plt.ylabel('x-value')
```

```python
x = np.arange(1,10, 0.001)
x
```

```python
plt.plot(x, x**2, label = 'Quadratic')
plt.plot(x, x**3, label = 'Cubic')
plt.plot(x, x*4, label = 'Linear')
plt.plot(x, 2**x, label ='Exponential')
plt.legend()
```

```python
plt.plot(x, x**2, x, x**3, x, 2**x)
```
